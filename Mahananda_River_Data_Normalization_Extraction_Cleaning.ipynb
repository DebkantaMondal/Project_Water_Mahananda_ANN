### Normalize Dataframes using MinMax Normalization

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
normal_df = normalize(df)
```

```python
normal_df.hist(figsize=(14,14))
plt.show()
```

```python
sns.distplot(normal_df['COD'])
plt.show()
```
